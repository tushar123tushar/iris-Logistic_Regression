```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
sns.heatmap(df.isnull() , yticklabels=False)
```

```python
df['species'].unique()
```

```python
df=df[df['species'] != 'setosa']
```

```python
df
```

```python
df['species'] = df['species'].map({'versicolor': 0  , 'virginica' : 1})
```

```python
df
```

```python
sns.heatmap(df.isnull() , yticklabels=False)
```

```python
#independent and Dependent Variable
df.head()
```

```python
x = df.iloc[: , : -1]
y = df.iloc[: , -1]
```

```python
x
```

```python
y
```

```python
#import train test split

from sklearn.model_selection import  train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
#Logistic 

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
Logistic_Regression = LogisticRegression()
```

```python
parameters  = {'penalty': ['elasticnet', 'l2', 'l1'] , 'C' : [1,2,3,4,5,6,10,5,450,102] , 'max_iter':[100,200,300]}
```

```python
class_fire = GridSearchCV(Logistic_Regression , param_grid=parameters , scoring='accuracy' , cv =5)
```

```python
class_fire.fit(X_train , y_train)
```

```python
print(class_fire.best_params_)
```

```python
print(class_fire.best_score_)
```

```python
#prediction

x_pred = class_fire.predict(X_test)
```

```python
x_pred
```

```python
#accuracy
from sklearn.metrics import r2_score , accuracy_score , classification_report
```

```python
score = accuracy_score(x_pred , y_test)
```

```python
score
```

```python
classification_report = classification_report(x_pred , y_test)
```

```python
classification_report
```
